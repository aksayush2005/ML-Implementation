<h3>Implementation of Ridge and Lasso algorithms using Scikit Learn</h3>

```python
#Importing essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
#importing the data
data=pd.read_csv("Boston-house-price-data.csv")
data.head()
```

```python
#To see essential information about the dataset
data.info()
```

```python
#to see various statistical data related to the given data
data.describe()
```

```python
#Would predict cost of a house using all the other features
#Splitting the dataset into independent and dependent features
x=data.iloc[:,:-1]
y=data.iloc[:,13]
x.head()
```

```python
#Splitting the dataset into independent and dependent features
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
#improving accuracy by standard scaler
from sklearn.preprocessing  import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#Implementation of Ridge Regression
from sklearn.linear_model import Ridge
model=Ridge(alpha=1)
#Using GridSearchCV for hyperparameter tuning
from sklearn.model_selection import GridSearchCV
params={"alpha":[1,5,10,100,200]}
cv=GridSearchCV(model,params)
#fitting the training data
cv.fit(x_train,y_train)
#predicting using the model
y_pred=cv.predict(x_test)
cv.best_estimator_
```

```python
#Representation of error in prediction using kde plot
sns.displot(y_test-y_pred,kind="kde")
plt.show()
```

```python
#Metrics using r2 scoring
from sklearn.metrics import r2_score
r2_score=r2_score(y_test,y_pred)
print(r2_score)
```

```python
#Implementing Lasso Algorithm
from sklearn.linear_model import Lasso
model=Lasso()
#Hyperparameter tuning
from sklearn.model_selection import GridSearchCV
params={"alpha":[1,5,10,100,200]}
cv=GridSearchCV(model,params)
#fitting the data
cv.fit(x_train,y_train)
y_pred=cv.predict(x_test)
print(cv.best_score_)
print(cv.best_estimator_)
#Scoring using r2 scoring
from sklearn.metrics import r2_score
r2_score=r2_score(y_test,y_pred)
print(r2_score)
```

```python
#Graph for Lasso Regression for deviation of predicted from actual
sns.displot(y_test-y_pred,kind="kde")
plt.show()
```
