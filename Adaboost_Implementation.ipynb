<h3>Adaboost Implementation</h3>

```python
#importing essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyplot
```

```python
#Importing Dataset
data=pd.read_excel("titanic3.xlsx")
data=pd.DataFrame(data)
data.head()
```

```python
#feature selection
data=data.drop(columns=["ticket","name","fare","body","boat"])
#dropping cabin as there are many empy values
data=data.drop(columns=["cabin","home.dest"])
```

```python
#Encoding the categorical features
data.sex=data.sex.map({"male":1,"female":0})
```

```python
data.describe()
```

```python
#Categorical encoding amd filling up null values
data.embarked=data.embarked.map({'S':0,'C':1,'Q':2})
data["embarked"]=data["embarked"].fillna(1)
```

```python
#filling up null values with median
data.age=data.age.fillna(28)
```

```python
#Splitting the data into independent and dependent features
x=data[["pclass","sex","age","sibsp","parch","embarked"]]
y=data["survived"]
#Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
#Scaling a data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#importing Adaboost
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
model=AdaBoostClassifier()
params={"n_estimators":[25,50],"learning_rate":[0.5,1]}
cv=GridSearchCV(model,params,cv=5)
cv.fit(x_train,y_train)
```

```python
cv.best_estimator_
```

```python
#Predicting using the model
y_pred=cv.predict(x_test)
```

```python
#Classification Report
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```
