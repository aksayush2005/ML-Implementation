<h3>Random Forest Implemetation</h3>

```python
#importing essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing Dataset
data=pd.read_csv("water_potability.csv")
data.head()
```

```python
data.describe()
```

```python
#Filling up the NA values using their median values
data.ph=data.ph.fillna(7.036752)
data.Sulfate=data.ph.fillna(333.073546)
data.Trihalomethanes=data.ph.fillna(66.622485)
```

```python
#Dividing the data into independent and dependent features
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
#Dividing the dataset into train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
#Scaling the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
#importing the random forest model
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
from sklearn.model_selection import GridSearchCV
params={"n_estimators":[50,100,150],"criterion":["gini","entropy"]}
cv=GridSearchCV(model,params,cv=5)
cv.fit(x_train,y_train)
y_pred=cv.predict(x_test)
```

```python
#Classification Report
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```
