<h3>Logistic Regression implementation using Scikit Learn</h3>

```python
#Importing essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import GridSearchCV
```

```python
#Importing the datatset
from sklearn.datasets import load_digits
x,y=sklearn.datasets.load_digits(as_frame=True,return_X_y=True)
data=pd.concat([x,y])
data.dropna(subset="target")
```

```python
#Importing the Logistic Regression model
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
#hyperparameter tuning
params={"penalty":["l1","l2"],"C":[1,2,3,4,5,0.5,0.8],"fit_intercept":[True,False]}
cv=GridSearchCV(model,params,cv=5)
#train,test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.33)
#fitting the data
cv.fit(x_train,y_train)
#predicting using model
y_pred=cv.predict(x_test)
print(y_pred)
#Classification report
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
```
